```python
import pandas as pd
content =pd.read_excel('douban_data.xlsx')
content.head(5)
```

# 将价格转化为元   将字转化为万

```python
# 创建新列保存带单位的价格
content['price'] = (content['fixedPrice']/100).round(2).astype(str) + '元'
content['wordCount_']= (content['wordCount']/10000).round(2).astype(str) + '万字'
# 验证结果
print(content[['fixedPrice', 'price','wordCount_','wordCount']])
```

```python
import numpy as np
# 选择所有数值型列（排除字符串格式列）
numeric_cols = content.select_dtypes(include=[np.number]).columns

# 对数值型数据进行范围限定
content[numeric_cols] = content[numeric_cols].clip(lower=-2.5, upper=2.5)

# 验证结果
print(content[numeric_cols].describe())
```

```python
data=pd.get_dummies(content,drop_first=True,dtype=float)
data
```

```python
import matplotlib.pyplot as plt

# 假设 dates 是一个表示时间的列表或数组，这里用 index 代替
dates = content.index

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# 图表1: fixedPrice 随时间变化的趋势图
ax1.plot(dates, content['fixedPrice'], 'r')
ax1.set_xlabel('Index'); ax1.set_ylabel('Fixed Price'); ax1.set_title('Fixed Price Trend')

# 图表2: wordCount 随时间变化的趋势图
ax2.plot(dates, content['wordCount'], 'b')
ax2.set_xlabel('Index'); ax2.set_ylabel('Word Count'); ax2.set_title('Word Count Trend')

# 图表3: fixedPrice 和 wordCount 的散点图
ax3.scatter(content['fixedPrice'], content['wordCount'])
ax3.set_xlabel('Fixed Price'); ax3.set_ylabel('Word Count'); ax3.set_title('Fixed Price vs Word Count')

# 图表4: shortName 类别下的 fixedPrice 分布直方图
content.groupby('shortName')['fixedPrice'].mean().plot(kind='bar', ax=ax4)
ax4.set_xlabel('Short Name'); ax4.set_ylabel('Average Fixed Price'); ax4.set_title('Average Fixed Price by Short Name')

plt.tight_layout(pad=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 假设 content 是你的 DataFrame
plt.figure(figsize=(10, 6))
plt.boxplot(content['fixedPrice'], vert=False)  # 设置为水平箱线图
plt.title('Box Plot of Fixed Price')
plt.xlabel('Fixed Price')
plt.yticks([1], ['fixedPrice'])  # 设置 y 轴标签
plt.grid(True)
plt.show()
```

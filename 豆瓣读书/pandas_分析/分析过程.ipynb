```python
import matplotlib.pyplot as plt


# 设置中文字体和负号显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

dates = content.index

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# 图表1: fixedPrice 随时间变化的趋势图
ax1.plot(dates, content['fixedPrice'], 'r')
ax1.set_xlabel('Index'); ax1.set_ylabel('Fixed Price'); ax1.set_title('Fixed Price Trend')

# 图表2: wordCount 随时间变化的趋势图
ax2.plot(dates, content['wordCount'], 'b')
ax2.set_xlabel('Index'); ax2.set_ylabel('Word Count'); ax2.set_title('Word Count Trend')

# 图表3: fixedPrice 和 wordCount 的散点图
ax3.scatter(content['fixedPrice'], content['wordCount'])
ax3.set_xlabel('Fixed Price'); ax3.set_ylabel('Word Count'); ax3.set_title('Fixed Price vs Word Count')

# 图表4: shortName 类别下的 fixedPrice 分布直方图
content.groupby('shortName')['fixedPrice'].mean().plot(kind='bar', ax=ax4)
ax4.set_xlabel('Short Name'); ax4.set_ylabel('Average Fixed Price'); ax4.set_title('Average Fixed Price by Short Name')

plt.tight_layout(pad=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# 生成月型数据
X, y = make_moons(n_samples=300, noise=0.15, random_state=42)

def plot_dataset(X, y, axes):
    """
    绘制数据集的散点图。

    参数:
        X: 特征矩阵
        y: 标签向量
        axes: 图像的坐标轴范围 [xmin, xmax, ymin, ymax]
    """
    # 绘制类别为0的数据点（蓝色方形）
    plt.plot(X[:, 0][y == 0], X[:, 1][y == 0], 'bs')
    # 绘制类别为1的数据点（绿色三角形）
    plt.plot(X[:, 0][y == 1], X[:, 1][y == 1], 'g^')
    # 设置坐标轴范围
    plt.axis(axes)
    # 显示网格线
    plt.grid(True, which='both')
    # 设置x轴标签
    plt.xlabel(r"$x_1$", fontsize=20)
    # 设置y轴标签并旋转90度
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

# 调用绘图函数并显示图像
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.svm import LinearSVC
from matplotlib.colors import ListedColormap
# 生成月型数据
X, y = make_moons(n_samples=300, noise=0.15, random_state=42)

# 定义并创建一个包含多项式特征、标准化和线性SVM分类器的管道
polynomial_svm_clf = Pipeline([
    ('poly_features', PolynomialFeatures(degree=3)),  # 添加多项式特征
    ('scaler', StandardScaler()),                     # 标准化特征
    ('svm_clf', LinearSVC(C=10, loss='hinge'))        # 线性SVM分类器
])

# 训练模型
polynomial_svm_clf.fit(X, y)

# 绘制决策边界函数
def plot_decision_boundary(clf, X, y, axes=[-1.5, 2.5, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if contour:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

# 绘制数据集和决策边界
plot_decision_boundary(polynomial_svm_clf, X, y)
plt.show()
```

```python
from sklearn.decomposition import PCA
polynomial_svm_clf.fit(X, y)

# 使用PCA将数据降到二维
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
def plot_reduced_data(clf, X_reduced, y, axes=[-1.5, 2.5, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if contour:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X_reduced[:, 0][y==0], X_reduced[:, 1][y==0], "bs")
    plt.plot(X_reduced[:, 0][y==1], X_reduced[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

# 创建图形
plt.figure(figsize=(12, 5))

# 子图1: 原始数据和决策边界
plt.subplot(121)
plot_decision_boundary(polynomial_svm_clf, X, y, [-1.5, 2.5, -1, 1.5])
plt.title("Original Data and Decision Boundary", fontsize=16)

# 子图2: 降维后的数据和决策边界
plt.subplot(122)
plot_reduced_data(polynomial_svm_clf, X_reduced, y, [-1.5, 2.5, -1, 1.5])
plt.title("Reduced Data and Decision Boundary", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
content = pd.read_excel('douban_data.xlsx')

# 示例数据（请根据实际情况替换）
X1D = content['fixedPrice'].values.reshape(-1, 1)
XK = content[['fixedPrice', 'wordCount']].values
yk = content['shortName'].map({'历史': 0, '其他': 1}).fillna(1).values  # 假设'历史'为0，其他为1

# 创建图形
plt.figure(figsize=(14, 6))

# 子图1: 原始数据分布
plt.subplot(121)
plt.gca().get_yaxis().set_ticks([0, 0.25, 0.5, 0.75, 1])
plt.scatter(X1D, yk, c=yk, cmap='viridis', alpha=0.5)
plt.xlabel(r"$fixedPrice$", fontsize=20)
plt.ylabel(r"Category", fontsize=14)
plt.annotate(r'$\mathbf{x}$',
             xy=(X1D[3, 0], yk[3]),
             xytext=(-0.5, 0.20),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.text(min(X1D) - 1000, 0.9, "历史", ha="center", fontsize=20)
plt.text(min(X1D) - 1000, 0.1, "其他", ha="center", fontsize=20)
plt.axis([min(X1D) - 1000, max(X1D) + 1000, -0.1, 1.1])

# 子图2: 变换后的数据分布
plt.subplot(122)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(XK[:, 0][yk==0], XK[:, 1][yk==0], "bs")
plt.plot(XK[:, 0][yk==1], XK[:, 1][yk==1], "g^")
plt.xlabel(r"$fixedPrice$", fontsize=20)
plt.ylabel(r"$wordCount$", fontsize=20, rotation=0)
plt.annotate(r'$\phi\left(\mathbf{x}\right)$',
             xy=(XK[3, 0], XK[3, 1]),
             xytext=(0.65, 0.50),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.axis([min(XK[:, 0]) - 1000, max(XK[:, 0]) + 1000, min(XK[:, 1]) - 10000, max(XK[:, 1]) + 10000])

plt.subplots_adjust(right=1)
plt.show()
```
